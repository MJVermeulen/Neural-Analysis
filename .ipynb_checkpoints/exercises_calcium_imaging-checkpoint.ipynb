```python
ROI_array = cm.load(input_tif_file_path)
N = 5
t,x_size,y_size = np.shape(ROI_array)

random_x,random_y = np.random.randint(x_size,size = N),np.random.randint(y_size,size =N)
pixels = np.vstack((random_x,random_y))

random_pixel_values = []

fig = plt.figure()
(ax1, ax2) = plt.subplots(2, sharex=False, sharey=False)

for i in range(N):
    one_pixel = []
    for j in range(t):
        #print(ROI_array[j,random_x[i],random_y[i]])
        one_pixel = np.append(one_pixel,ROI_array[j,random_x[i],random_y[i]])
    ax = plt.subplot(211)
    ax.plot(one_pixel,label = f'Location {random_x[i],random_y[i]}')
    ax.set_xlabel('time')
    ax.set_ylabel('pixel value')
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    ax.set_title('Pixel value for random pixels over time')
    ax = plt.subplot(212)
    plt.hist(one_pixel, edgecolor='None', alpha = 0.5)
    #ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, shadow=True, ncol=5)   
    #ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    ax.set_xlabel('pixel value')
    ax.set_ylabel('frequency')
    ax.set_title('histogram')

```

The difference in the regions is the mean pixel value. We can also see that neurons have a wider distribution. Neurons have a wider range of pixel value since it turns on and off. For all pixels we see that  there is a rise in pixel values at around t = 40

We need to segment the neurons because individual pixels can behave stangly, there can be something wrong with this particular pixel in the setup. Also the image may be shaking. If the pixel is a neuron it most likily also has lit up pixels as neighbours. The shaking will than not be noticed and the temporal tracing will be completely wrong. Even if the neuron is moving the value of the pixel doesn't change since it just measures the a neighbour with the same value.
