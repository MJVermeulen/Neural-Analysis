```python
#Import the data
download_data('https://surfdrive.surf.nl/files/index.php/s/7BosHgemAseANIc')
with open('data/moving_lfp.pickle', 'rb') as handle:
    lfp_data = pickle.load(handle)

lfp = lfp_data['lfp']
sampling_rate = lfp_data['sampling_rate']
print(f'Loaded {len(lfp)/sampling_rate} seconds of LFP, sampled at {sampling_rate} Hz')
print(lfp)
```

```python
lfp_data
```

```python
#Take a few seconds of the data, and filter them in different frequency bands.
lfp_data_reduced = lfp[int(100*sampling_rate):int(110*sampling_rate)] #Take data from the 100 to 110 seconds of the measurments
from scipy.fft import fft, fftfreq

from scipy.signal import butter
from scipy.signal import sosfilt


lfp_data_reduced -= np.mean(lfp_data_reduced)


filter_center = [10,25,45,65,85,105,125,145,165]
filter_offset = 5
filter_offset_array = [5,10,15,20,25,30,40,50]
filter_center_fixed = 51
#Keep the center fixed but the width changes

plot_offset = 0

plt.figure(figsize=(15,5))
for i in filter_offset_array:
    plot_bandpass_lfp = lf.bandpass_filter(lfp_data_reduced,low_f= filter_center_fixed-i,high_f= filter_center_fixed +i, sampling_rate=sampling_rate)
    plot_offset -= 500 #to plot the signal on different lines
    plt.plot(plot_bandpass_lfp[:2000]+plot_offset,label= f'Bandwidth ({filter_center_fixed-i} - {filter_center_fixed+i} Hz) ' )

plt.axis('off')
plt.legend(loc=(0,1))
```

We keep the center frequency fixed but increase the bandwidth, we see that as we increase the bandwidth the signal becomes more 'random'. This is logical since we keep including more frequencies and thus more wave functions. Thus the combined wavefunction becomes more 'random'. It also looks like the signal is shifting to the right. 

```python
#Keep the width fixed but the center changes
plot_offset = 0

plt.figure(figsize=(15,5))
for i in filter_center:
    plot_bandpass_lfp = lf.bandpass_filter(lfp_data_reduced,low_f= i - filter_offset,high_f= i + filter_offset, sampling_rate=sampling_rate)
    plot_offset -= 200 #to plot the signal on different lines
    plt.plot(plot_bandpass_lfp[:2000]+plot_offset,label= f'Bandwidth ({i-filter_offset} - {i+ filter_offset} Hz) ' )

plt.axis('off')
plt.legend(loc=(0,1))
```

If we keep the bandwidth fixed but change the position of the centre of the frequency all the graphs look different. As the centre frequency is increased the amplitude of the signal decreases. We can also see amplitude burst at all the frequencies except for 5-15 hz. The burst don't show at pattern with respect tot the other frequency centers. 

```python
#Filter order
import time
filter_order = [1,2,5,10,25,50] #the higher the order, the sharper the filter frequency response, 
                 #but with more computations and potental numerical instability
low_f = 20
high_f = 100

plot_offset = 0
time_array = []


plt.figure(figsize=(10,5))
plt.plot(np.linspace(0,2,2000),lfp[:2000],label='raw signal')
for i in filter_order:
    
    plot_offset += 400
    #define filter object
    filter = butter(i,[low_f,high_f],
                    btype='band', 
                    output='sos', 
                    fs=sampling_rate)
                
    #apply filter to lfp  
    st = time.time()
    wideband_lfp = sosfilt(filter, lfp)
    et = time.time()
    time_array.append(et-st)
    plt.plot(np.linspace(0,2,2000),wideband_lfp[:2000] - plot_offset,label = f'filter order = {i}, time = {et - st}')
    
    
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.legend()
plt.axis('off')
plt.legend(loc=(0,1))
```

```python
plt.plot(filter_order,time_array)
plt.xlabel('filter order')
plt.ylabel('time (s)')
plt.title('time it takes to calculate for each filter order')
```

Increasing the filter order should increase the sharpness of the signal. A filter order higher than one increases the sharpness however a filter order higher than 2 does not seem to improve the sharpness that much. There also seem to be a shift to the right. At lower filter order there does not seem to be a difference in the tame it takes to compute but at higher filter order >10 the time it takes to calcutlate increases drastically.

In this exercises you will explore the difference between real and complex wavelets in the continous wavelet transform.
- Import the data form `data/moving_lfp.pickle`
- Select a period of a few seconds of the signal and compute the spectrogram. 
  Try the wavelet transform with wavelet `mexh` (mexican hat),`gaus` (gaussian) and `morl` (morlet). These are all real wavelets.
- What differences do you see with respect to the spectrum computed with the complex Morlet tranform ? Comment on why do you think this might be the case.

```python
from scipy import signal
from pywt import scale2frequency
from pywt import cwt
```

```python
#Plot the spectogram
start = 100
end = 105

lfp_data_reduced = lfp[int(start*sampling_rate):int(end*sampling_rate)] #Take data from the 100 to 110 seconds of the measurments
lfp_data_reduced -= np.mean(lfp_data_reduced)
wavelet_types = ['gaus1','mexh', 'morl']
k =1 

for i in wavelet_types:
    
    frequencies =  np.linspace(0.1,20,100)/sampling_rate
    wavelet = i
    scales = scale2frequency(wavelet, frequencies)
    cwtmatr, _ = cwt(lfp_data_reduced, wavelet=wavelet, scales=scales,
                      sampling_period=1.0/sampling_rate)
    plt.figure(figsize=(10,15))
    plt.subplot(3,1,k)
    k +=1
    plt.imshow(abs(cwtmatr)**2, extent=[start,end,frequencies[-1]*sampling_rate,
                             frequencies[0]*sampling_rate], aspect='auto', cmap='CMRmap')
    plt.title(f'{i}')


    plt.xlabel('time (s)')
```

The mexican hat and the gaus wavelet transform have very simular spectra. The morlet wave has a differnt spectrum. The morlet wavelet transform has a simular locations however a lower value. This is because it includes the complex part of the wave which decreases its value.

## Exercise 3: REM sleep detection
---
Mammalian sleep is organized in different phases. [Rapid Eye Movement (REM) sleep](https://en.wikipedia.org/wiki/Rapid_eye_movement_sleep) is a well studied phase, with clear markers that make it detectable from LFP data. One of the most used marker is the ratio (or difference) between the oscillatory power in the theta and delta bands in the hippocampus. During REM sleep, the former dominates, in a pattern that resembles active behaviour.  
In this exercise we will use the tools we learned about to detect and inspect the REM and non-REM (nREM) phases of sleep in data recorded from the hippocampus of a sleeping mouse.

- Import the data form `data/moving_lfp.pickle`
- Compute the instantaneous power in the theta (6-10 Hz) and delta (2-4 Hz) frequency bands, for the whole signal.
- z-score the two power signals (subtract the mean and divide by the standard deviation, you can use `scipy.stats.zscore()`)
- Compute and plot the power difference (`theta_power - delta_power`) for the sleep session.
- Compute the REM (power difference $>0.2$) and nREM (power diffewrence $<0.2$) sleep periods.
- Plot examples of REM vs nREM lfp signals (wideband filtered betw. 1 and 200 Hz to get rid of some of the noise).
- Explore how changing the threshold affects the amount of REM sleep detected.

```python
start = 0#in seconds
end = 600.133 # in seconds

theta_lfp = lf.bandpass_filter(lfp,low_f=6,high_f=10, sampling_rate=sampling_rate)
delta_lfp = lf.bandpass_filter(lfp,low_f=2,high_f=4, sampling_rate=sampling_rate)

sig_theta = theta_lfp[int(start*sampling_rate):int(end*sampling_rate)]
sig_theta = sig_theta - np.mean(sig_theta) # get rid of 0 frequency component

sig_delta = delta_lfp[int(start*sampling_rate):int(end*sampling_rate)]
sig_delta = sig_delta - np.mean(sig_delta) # get rid of 0 frequency component

frequencies_theta =  np.linspace(6,10,50)/sampling_rate
frequencies_delta =  np.linspace(2,4,50)/sampling_rate

wavelet = 'cmor1.0-1.0'
scales_theta = scale2frequency(wavelet, frequencies_theta)
scales_delta = scale2frequency(wavelet, frequencies_delta)

cwtmatr_theta, _ = cwt(sig_theta, wavelet=wavelet, scales=scales_theta,
                  sampling_period=1.0/sampling_rate)

cwtmatr_delta, _ = cwt(sig_delta, wavelet=wavelet, scales=scales_delta,
                  sampling_period=1.0/sampling_rate)

power_theta = np.mean(abs(cwtmatr_theta)**2,axis=0)
power_delta = np.mean(abs(cwtmatr_delta)**2,axis=0)

plt.tight_layout()
plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
plt.plot(np.linspace(start,end,len(power_theta)),power_theta)
plt.xlabel('time (s)')
plt.ylabel('power (a.u.)')
plt.title('theta')


plt.subplot(2,1,2)
plt.plot(np.linspace(start,end,len(power_delta)),power_delta)
plt.xlabel('time (s)')
plt.ylabel('power (a.u.)')
plt.title('delta')
```

```python
#Compute the z-score
from scipy import stats

theta_zscore = stats.gzscore(power_theta, axis=0, ddof=0, nan_policy='propagate')
delta_zscore = stats.zscore(power_delta, axis=0, ddof=0, nan_policy='propagate')
#print(theta_zscore,delta_zscore)
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(np.linspace(start,end,len(theta_zscore)),theta_zscore, label = 'theta')
plt.xlabel('time (s)')
plt.ylabel('power (a.u.)')
plt.title('zscore theta')
plt.legend()

plt.subplot(2,1,2)
plt.plot(np.linspace(start,end,len(delta_zscore)),delta_zscore,label = 'delta')
plt.xlabel('time (s)')
plt.ylabel('power (a.u.)')
plt.title('zscore delta')
plt.legend()
```

```python
#plot the power difference
diff_power = theta_zscore - delta_zscore

plt.figure(figsize = (10,5))
plt.plot(np.linspace(start,end,len(diff_power)),diff_power)
#plt.xlim([0,12]) # restrict to 
#plt.xlabel('Hz')
plt.xlabel('time (s)')
plt.ylabel('power (a.u.)')
plt.title("power difference, theta- delta")
plt.legend()
```

```python
#Plot the REM and nREM sleep period
from scipy.signal import butter
from scipy.signal import sosfilt

threshold = 0.2
REM = diff_power > threshold
nREM = diff_power < threshold


filter_order = 5 #the higher the order, the sharper the filter frequency response, 
                 #but with more computations and potental numerical instability
low_f = 1
high_f = 200

#define filter object
filter = butter(filter_order,[low_f,high_f],
                btype='band', 
                output='sos', 
                fs=sampling_rate)
    
wideband_lfp_REM = sosfilt(filter, lfp)
#wideband_lfp_REM = wideband_lfp[:len(diff_power)]

REM_array = np.empty_like(wideband_lfp_REM)
REM_array[REM] = wideband_lfp_REM[REM]

    
wideband_lfp_nREM = sosfilt(filter, lfp)
#wideband_lfp_nREM = wideband_lfp[:len(diff_power)]

nREM_array = np.empty_like(wideband_lfp_nREM)
nREM_array[nREM] = wideband_lfp_nREM[nREM]


```

```python
#Both REM an NRem
start = 300000 
end = 305000
REM_array_plot = REM_array[start:end]
nREM_array_plot = nREM_array[start:end]
#print(REM_array_plot)

plt.figure(figsize = (10,5))
plt.plot(np.linspace(start,end,len(REM_array_plot))/sampling_rate,REM_array_plot, label = 'REM')
plt.plot(np.linspace(start,end,len(nREM_array_plot))/sampling_rate,nREM_array_plot, label = 'nREM')
plt.xlabel('time (s)')
plt.ylabel('power (a.u.)')
plt.legend()
```

We can see that the REM and nREM sleep follow simular patterns when active
